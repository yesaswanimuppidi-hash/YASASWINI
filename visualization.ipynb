```python
df["Timestamp"] = pd.to_datetime(df["Timestamp"], errors="coerce")
```

```python
df = pd.read_excel("LIBRARY DATA SET.xlsx")
print(df.shape)
print(df.columns)
df.head()
```

Load Dataset

```python
category_counts = df["Category"].value_counts()

plt.figure(figsize=(8,5))
plt.bar(category_counts.index, category_counts.values)
plt.title("Books Count by Category")
plt.xlabel("Category")
plt.ylabel("Number of Books")
plt.xticks(rotation=45)
plt.show()
```

```python
top_books = df["Title"].value_counts().head(10)

plt.figure(figsize=(10,5))
plt.bar(top_books.index, top_books.values)
plt.title("Top 10 Most Issued Books")
plt.xlabel("Book Title")
plt.ylabel("Issue Count")
plt.xticks(rotation=90)
plt.show()
```

Returned vs Not Returned (Pie Chart)

```python
returned = df[df['Status'] == 'Present'].shape[0]
not_returned = df[df['Status'].isin(['Checked Out', 'Missing'])].shape[0]

plt.figure(figsize=(6,6))
plt.pie([returned, not_returned], labels=["Returned", "Not Returned"], autopct="%1.1f%%")
plt.title("Returned vs Not Returned Books")
plt.show()
```

Monthly Book Issues (Line Chart)

```python
df["Issue_Month"] = df["Timestamp"].dt.month
monthly_issues = df["Issue_Month"].value_counts().sort_index()

plt.figure(figsize=(8,5))
plt.plot(monthly_issues.index, monthly_issues.values, marker="o")
plt.title("Monthly Book Issue Trend")
plt.xlabel("Month")
plt.ylabel("No. of Issues")
plt.xticks(range(1,13))
plt.grid(True)
plt.show()
```

```python
import numpy as np

# Create a 'Fine_Amount' column with random values for demonstration purposes
df['Fine_Amount'] = np.random.randint(0, 50, size=len(df))

print("Added 'Fine_Amount' column to the DataFrame with random values.")
display(df.head())
```

```python
df['Issue_Date'] = df['Timestamp']
# Generate 'Return_Date' by adding a random number of days (1-30) to 'Issue_Date'
df['Return_Date'] = df['Issue_Date'] + pd.to_timedelta(np.random.randint(1, 30, size=len(df)), unit='D')

print("Added 'Issue_Date' and 'Return_Date' columns to the DataFrame.")
display(df.head())
```

```python
plt.figure(figsize=(8,5))
plt.hist(df["Fine_Amount"], bins=10)
plt.title("Fine Amount Distribution")
plt.xlabel("Fine Amount")
plt.ylabel("Frequency")
plt.show()
```

```python
df["Days_Taken"] = (df["Return_Date"] - df["Issue_Date"]).dt.days

plt.figure(figsize=(8,5))
plt.scatter(df["Days_Taken"], df["Fine_Amount"])
plt.title("Days Taken vs Fine Amount")
plt.xlabel("Days Taken")
plt.ylabel("Fine Amount")
plt.grid(True)
plt.show()
```
