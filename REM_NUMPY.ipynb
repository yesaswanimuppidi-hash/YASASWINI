Import Libraries

```python
import pandas as pd
import numpy as np
```

Load Excel File

```python
df = pd.read_excel("LIBRARY DATA SET.xlsx")
print("Original Shape:", df.shape)
df.head()
```

Convert DataFrame to NumPy Array

```python
data = df.to_numpy()
print("NumPy Data Shape:", data.shape)
```

Check Duplicate Rows Count (NumPy)

data_str = data.as

```python
data_str = data.astype(str)

unique_rows, counts = np.unique(data_str, axis=0, return_counts=True)

duplicate_rows = np.sum(counts > 1)

print("No of Duplicate Rows:", duplicate_rows)
```

Convert Back to DataFrame


```python
df_unique = pd.DataFrame(unique_rows, columns=df.columns)
df_unique.head()
```
