**REMOVING DUPLICATES**

Import Libraries

```python
import pandas as pd
```

Load Excel File

```python
df = pd.read_excel("/content/sample_data/LIBRARY DATA SET.xlsx")
print(df.shape)
df.head()
```

Check Duplicate Rows

```python
duplicates = df.duplicated()
print(duplicates.sum())
```

View Duplicate Records

```python
df[df.duplicated()]
```

Remove Full Duplicate Rows

```python
df_clean = df.drop_duplicates()
print(df_clean.shape)
```

Save Clean Data

```python
df_clean.to_excel("LIBRARY_DATA_CLEAN.xlsx", index=False)
print("Duplicates removed and file saved successfully!")
```
